```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
```

```python
data_file = "NYPD_Shooting_Incident_Data__Historic_.csv"

df = pd.read_csv(data_file)
df.head()
```

```python
with sns.axes_style('whitegrid'):
    g = sns.factorplot("YEAR", data=df, aspect=1.5,
                       kind="count", color='steelblue')
    g.set_xticklabels(step=1)
    plt.title("Shootings Per Year in NYC")
    plt.ylabel("COUNT OF SHOOTING")
    plt.xlabel("YEAR")
    
g.savefig("Shootings Per Yr in NYC")
```

```python
with sns.axes_style('whitegrid'):
    g = sns.factorplot("VIC_AGE_GROUP", data=df, aspect=1.,
                       kind="count", color='coral')
    g.set_xticklabels(step=1)
    plt.title("VICTIM AGE RANGE: 2006-2018")
    plt.ylabel("COUNT OF SHOOTING")
    plt.xlabel("VICTIM AGE RANGE")
g.savefig("Victim age rng")
```

```python
with sns.axes_style('whitegrid'):
    g = sns.factorplot("VIC_RACE", data=df, aspect=2.9,
                       kind="count", color='BLUE')
    g.set_xticklabels(step=1)
    plt.title("VICTIM by RACE: 2006-2018")
    plt.ylabel("COUNT OF SHOOTING")
    plt.xlabel("VICTIM RACE")
g.savefig("Victim by Race:2006-2018")
```

```python
with sns.axes_style('whitegrid'):
    g = sns.factorplot("VIC_SEX", data=df, aspect=.5,
                       kind="count", color='BLUE')
    g.set_xticklabels(step=1)
    plt.title("VICTIM SEX: 2006-2018")
    plt.ylabel("COUNT OF SHOOTING")
    plt.xlabel("VICTIM SEX")
g.savefig("VIctim Sex")
```

```python
with sns.axes_style('whitegrid'):
    g = sns.factorplot("BORO", data=df, aspect=1.5,
                       kind="count", color='BLUE')
    g.set_xticklabels(step=1)
    plt.title("Shootings by Borough: 2006-2018")
    plt.ylabel("COUNT OF SHOOTING")
    plt.xlabel("BOROUGH")
g.savefig("Shootings by Boro")
```

```python
with sns.axes_style('whitegrid'):
    g = sns.factorplot("YEAR", data=df, aspect=3.0, kind='count', hue='BORO', order=range(2006, 2019))
    g.set_ylabels('Number of shootings')
    plt.title("Shootings by Boro YOY")
g.savefig("Shootings by Boro YOY")
```

```python
with sns.axes_style('whitegrid'):
    g = sns.factorplot("YEAR", data=df, aspect=1.8, kind='count', hue='VIC_SEX', order=range(2006, 2019))
    g.set_ylabels('Number of shootings')
    plt.title("Victims by Sex YOY")
g.savefig("VIC SEX YOY")
```

```python
with sns.axes_style('whitegrid'):
    g = sns.factorplot("YEAR", data=df, aspect=2.0, kind='count', hue='VIC_AGE_GROUP', order=range(2006, 2019))
    g.set_ylabels('Number of shootings')
    plt.title("VICTIM by Age Group YOY")
g.savefig("VIC AGE YOY")
```

```python
with sns.axes_style('whitegrid'):
    g = sns.factorplot("YEAR", data=df, aspect=2.0, kind='count', hue='VIC_RACE', order=range(2006, 2019))
    g.set_ylabels('Number of shootings')
    plt.title("VICTIM Race YOY")
g.savefig("VIC RACE YOY")
```
