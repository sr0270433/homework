```python
# Dependencies
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
```

```python
data_file = "NYC_SHOOTING_BY_YEAR.csv"
data_file2 = "NYPD_Shooting_Incident_Data__Historic_.csv"
df = pd.read_csv(data_file)
df2 = pd.read_csv(data_file2)
df.head(20)
```

```python
year = df.iloc[:, 0]
df2.head()
```

```python
number_incidents = df.iloc[:, 1]
vc_slope, vc_int, vc_r, vc_p, vc_std_err = stats.linregress(
    year, number_incidents)
vc_fit = vc_slope * year + vc_int
```

```python

fig, (ax1) = plt.subplots(1, sharex=True)
fig.suptitle("Number of Shootings YoY", fontsize=16, fontweight="bold")

ax1.set_xlim(min(year), max(year))
ax1.plot(year, number_incidents, linewidth=1, marker="o")
ax1.plot(year, vc_fit, "b--", linewidth=1)
ax1.set_ylabel("Number of Shootings")
```
